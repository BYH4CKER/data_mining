```python
#veri çerçevesindeki "Currency" sütununda bulunan 'Euro (EUR)' değerini 'Euro' ile değiştirmek için kullanılır.
df['Currency'] = df['Currency'].replace('Euro (EUR)', 'Euro')
```

```python
#//Verilerimizi Görselleştirme ve Grafikler Aracılığıyla Keşfetme
#plotly.express kütüphanesi kullanılarak df
fig=px.pie(df,names="Country",title="Ülke Dağılımı")
fig.show()
#veri çerçevesindeki "Country" sütununa dayalı bir pasta grafik oluşturur.
#names="Country": Pasta dilimlerini oluşturmak için "Country" sütunundaki benzersiz değerleri kullanır.
#title="Country Distribution": Grafiğin başlığı olarak "Country Distribution" ayarlanır.
```

```python
#plotly.express kütüphanesi kullanılarak df
fig=px.pie(df,names="Cost of Living",title="Yaşam Maliyeti Dağıtımı")
fig.update_traces(textposition="inside",textinfo="percent+label")
fig.show()
#veri çerçevesindeki "Cost of Living" sütununa dayalı bir pasta grafik oluşturur
#fig.update_traces(textposition="inside", textinfo="percent+label"):
#Dilimlerin üzerine hem yüzde hem de etiketlerin yazılmasını ve metinlerin dilimlerin içinde yer almasını sağlar.
```

```python
#seaborn kütüphanesi kullanılarak df
sns.countplot(data=df,x='Yaşam Maliyeti Dağıtımı')
#veri çerçevesindeki "Cost of Living" sütununun dağılımını görselleştiren bir sütun grafiği (bar plot) oluşturur.
#sns.countplot(data=df, x='Cost of Living'): df veri çerçevesindeki "Cost of Living" sütunundaki her benzersiz değerin kaç kez geçtiğini sayar ve
#bu bilgiyi bir sütun grafiği olarak görselleştirir.
#Sütun grafiği, her bir "Cost of Living" kategorisinin veri setindeki frekansını göstermeye yarar.
```

```python
# Sütunlara göre gruplandır
df.groupby('Cost of Living').size()
#veri çerçevesindeki "" sütununa göre gruplar oluşturur ve her grubun satır sayısını döndürür
#Bu, her "" kategorisinin veri setinde kaç kez geçtiğini gösteren bir çıktı üretir.
```

```python
#seaborn kütüphanesini kullanarak df

plt.figure(figsize=(12, 6))
sns.countplot(x='Country', hue='Cost of Living', data=df)
plt.title('Cost of Living in Different Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Cost of Living')
plt.show()
#veri çerçevesindeki "Country" sütununu x ekseninde ve "Cost of Living" sütununu bir kategori olarak gösteren bir sütun grafiği oluşturur.
#Bu grafik, her ülke için yaşam maliyeti kategorilerinin dağılımını görselleştirir.

#x='Country': X eksenine "Country" sütununu yerleştirir.
#hue='Cost of Living': Farklı yaşam maliyeti kategorilerini farklı renklerle gösterir.
#plt.xticks(rotation=90): Ülke isimlerini daha iyi görmek için x eksenindeki etiketleri 90 derece döndürür.
#plt.legend(title='Cost of Living'): Grafikteki renklerin neyi temsil ettiğini gösteren bir başlıklı açıklama ekler.
```

```python
df.groupby('Cost of Living').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
#veri çerçevesinde "Cost of Living" sütununa göre gruplandırılmış frekansları gösteren yatay bir sütun grafiği (bar chart) oluşturur
#ve renkleri sns.palettes.mpl_palette('Dark2') paletinden alır.
#Ayrıca, grafikte üst ve sağ kenarlıkların görünürlüğü kaldırılarak daha temiz bir görselleştirme sağlanır.
#Bu tür bir grafik, her "Cost of Living" kategorisinin veri setindeki frekansını ve oranını kolayca görmenizi sağlar.
```
